```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dataset.csv")
df.head()
```

```python
df["Calificación Final (Y)"].agg(["mean", "std"]).reset_index()
```

```python
df.groupby('Grupo (X)')['Calificación Final (Y)'].agg(['mean', 'std']).reset_index()
```

```python
df.groupby(["Perfil SRL","Grupo (X)"])["Calificación Final (Y)"].agg(["mean", "std"]).reset_index()
```

```python
sns.set_theme(style="whitegrid")

# Create the Boxplot
plt.figure(figsize=(8, 6))
ax = sns.boxplot(x='Grupo (X)', y='Calificación Final (Y)', hue="Grupo (X)", data=df, palette='Set2')

# Add titles and labels for the research context
plt.title('Comparación de Calificaciones: Aula Invertida vs. Tradicional', fontsize=14)
plt.xlabel('Metodología (Variable Independiente X)', fontsize=12)
plt.ylabel('Calificación Final (Variable Dependiente Y)', fontsize=12)

# Show the plot
plt.show()
```
